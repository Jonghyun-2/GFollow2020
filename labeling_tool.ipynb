```python
# dicom path를 입력으로 받아서 img반환 (ndarray)
# normalzize 미포함(mask 이미지)
def dcm2img_uint8(dcm_path):
    image_bytes = tf.io.read_file(dcm_path)
    
    decode_img_lossy = tfio.image.decode_dicom_image(image_bytes,scale='auto',on_error='lossy', dtype=tf.uint8)
    #############################################################################################
    # args
    # scale : datatype 변형시 fill stretching 
    # on_error : dtype uint 16 -> 8할때 error 방지
    #############################################################################################
    decode_img_lossy_ndr = decode_img_lossy.numpy()
    decode_img_lossy_ndr = decode_img_lossy_ndr[0,...,0]
    #decode_img_lossy_ndr_norm = decode_img_lossy_ndr[0,...,0]/255
    #decode_img_lossy_ndr_norm_add_axis = decode_img_lossy_ndr_norm[np.newaxis,...,np.newaxis]
    return decode_img_lossy_ndr
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import glob
import pydicom as pdi
import tensorflow_io as tfio
import tensorflow as tf
#paths

ROOT_TEST = r"E:\DataSet\CBIS-DDSM\test"
ROOT_TRAIN = r"E:\DataSet\CBIS-DDSM\train"
ROOT_VAL = r"E:\DataSet\CBIS-DDSM\val"

OUTPUT_TEST = r"E:\DataSet\CBIS-DDSM\Crop_test"
OUTPUT_TRAIN = r"E:\DataSet\CBIS-DDSM\Crop_train"
OUTPUT_VAL = r"E:\DataSet\CBIS-DDSM\Crop_val"

FULL_DCM_PATH_TRAIN = r'D:\Gproject\DataSet\tcia_data\CBIS-DDSM\*Train*\*\*full*\1-1.dcm'
FULL_DCM_PATH_TEST = r'D:\Gproject\DataSet\tcia_data\CBIS-DDSM\*Test*\*\*full*\1-1.dcm'

```

```python
# dicom path를 입력으로 받아서 img반환 (ndarray)
# normalzize 미포함(mask 이미지)
def dcm2img_uint8(dcm_path):
    image_bytes = tf.io.read_file(dcm_path)
    
    decode_img_lossy = tfio.image.decode_dicom_image(image_bytes,scale='auto',on_error='lossy', dtype=tf.uint8)
    #############################################################################################
    # args
    # scale : datatype 변형시 fill stretching 
    # on_error : dtype uint 16 -> 8할때 error 방지
    #############################################################################################
    decode_img_lossy_ndr = decode_img_lossy.numpy()
    decode_img_lossy_ndr = decode_img_lossy_ndr[0,...,0]
    
    return decode_img_lossy_ndr
```

```python
# test --> Crop_test
dcm_paths = glob.glob(FULL_DCM_PATH_TEST)
output_path = r'D:\Gproject\DataSet\tcia_data\dataset_dir\val'

len(dcm_paths)

def mouse_callback(event, x, y, flags, param):
    global now_x,now_y,ldown_x, ldown_y,lup_x, lup_y
    if event ==cv.EVENT_MOUSEMOVE:
        now_x,now_y = x,y
    if event == cv.EVENT_LBUTTONDOWN:
        ldown_x, ldown_y = x,y
    if event == cv.EVENT_LBUTTONUP:
        lup_x,lup_y = x,y
        print(ldown_y,lup_y)

now_x,now_y,ldown_x, ldown_y,lup_x, lup_y = -1,-1,-1,-1,-1,-1


for i in range(len(dcm_paths)):
    # 파일명 slicing
    dicom_mask_path_ID = dcm_paths[i].split('\\')[5]
    
    # dcm -> img(ndrrray)
    img = dcm2img_uint8(dcm_paths[i])
    ##################################################
    if img.shape[0] < 2000:
        print("*Changed*")
        str_list = dcm_paths[i].split('\\')
        str_list.pop()
        str_list.append('1-1.dcm')
        dcm = str_list[0]+"\\" + str_list[1]+"\\" + str_list[2]+"\\" + str_list[3]+"\\"  + str_list[4] +"\\" + str_list[5]+"\\" + str_list[6]+"\\" + str_list[7]+"\\" + str_list[8] 
        #print("Src path :", dcm)
        img = dcm2img_uint8(dcm_paths[i])
        #print(img.shape)
    ##################################################
    
    # 저장파일명
    output_name = output_path + "\\"+ dicom_mask_path_ID + "_full"+ ".png"
    #plt.imsave(output_name, img,cmap = 'gray')
    
    full_name = output_name
    cv.namedWindow(full_name,cv.WINDOW_NORMAL|cv.WINDOW_KEEPRATIO)
    cv.setMouseCallback(full_name,mouse_callback)
    cv.resizeWindow(full_name,500,800)
    cv.moveWindow(full_name,100,100)
    while True:
        #full_name = full_imgs[i]
        full_img = img
        width = full_img.shape[0]
        full_img_show = full_img.copy()
        cv.line(full_img_show,(0,now_y),(width,now_y),(255,0,0),10)
        cv.imshow(full_name, full_img_show)
        
        key = cv.waitKey(200)
        
        if key == 32:
            cv.destroyAllWindows()
            full_name_split = full_name.split("\\")[-1]
            
            # save cropped img(Full img)
            cv.imwrite(OUTPUT_TEST+"\\"+full_name_split, full_img[ldown_y:lup_y,...],params = [cv.IMWRITE_PNG_COMPRESSION,2])
            
            
            mask_names = full_name_split[:-8] + "*_mask.png" 
            mask_names = glob.glob(ROOT_TEST + "\\" + mask_names)
            for j in range(len(mask_names)):
                
                mask_name = mask_names[j]
                print(mask_name)
                mask_name_split = mask_name.split("\\")[-1]
                mask_img = cv.imread(mask_name,cv.IMREAD_UNCHANGED)
                # save cropped img(Mask img)
                cv.imwrite(OUTPUT_TEST+"\\"+mask_name_split, mask_img[ldown_y:lup_y,...],params = [cv.IMWRITE_PNG_COMPRESSION,2])
            break
        if key == 27:
            cv.destroyAllWindows()
            break
    if key ==27:
        break
```
