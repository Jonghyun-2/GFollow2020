```python
import pydicom as pdi
import numpy
import glob
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
def drawer(img):
    plt.figure(figsize=(16,8))
    plt.imshow(img,cmap='gray')
    
```

```python
# dicom image중 full mammogram images만 찾기

full_path = glob.glob('../../DataSet/tcia_data/CBIS-DDSM/*/*/*/*.dcm')

full_mammogram_images = []
ROI_mammogram_images = []
for i in range(len(full_path)):
    split = full_path[i].split("\\")
    fullORRoi = split[-2].split("-")
    if fullORRoi[-2] == "full mammogram images":
        full_mammogram_images.append(full_path[i])
    else:
        ROI_mammogram_images.append(full_path[i])
```

```python
test = full_mammogram_images[3]
test_r = ROI_mammogram_images[8]
print(test,'\n',test_r)
```

```python
a = pdi.read_file(test)
b = pdi.read_file(test_r)
```

```python
img = a.pixel_array
img_r = b.pixel_array

drawer(img)
```

```python
drawer(img_r)
```

```python
asdf = img.astype(np.uint8)
```

```python
asdf_ = img_r.astype(np.uint8)
```

```python
ans = cv.matchTemplate(asdf,asdf_,cv.TM_SQDIFF)
```

```python
plt.imshow(asdf,cmap='gray')
```

```python
a = cv.minMaxLoc(ans)
```
